```


######This is a program to solve x-or using neural network###############

import numpy as np
from keras.layers import Dense, LeakyReLU
from keras.models import Sequential
```

Input for our x-or problem is a 4x2 matrix

```
x = np.array([[0,0],[0,1],[1,0],[1,1]])
```

Outpur corresponding to given input is a 4x1 matrix

```
y = np.array([1,0,0,1])
```

Now we will define our model. It is going to be a simple sequential model with 2 layers ,1 hidden and 1 output layer.

```
model =  Sequential()
```

Hidden layer has 2 units and uses ReLu as activation. The input to hidden unit is 4 examples each having 2 features. ie a 4x2 matrix. Hence the dimensions of associated weight matrix would be 2x2.

```
act = LeakyReLU(alpha=0.3)
model.add(Dense(units=2,input_dim=2))
model.add(act)
```

Output layer has 1 unit and activation function used is sigmoid as we are dealing with two class classification problem. Sigmoid transforms it's input into [0,1] range

```
model.add(Dense(units=1,activation="sigmoid"))
```

Now, let's summarize our mode

```
model.summary()
```

Let's check what are the initial weights

```
print(model.get_weights());
```

Now we will compile our model, we will define our loss function as binary_crossentropy, we are using adam optimizer with metrics as accuracy i.e. our models focus it to increas accuracy.

```
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

Let's start training for our model. We will run 2000 interations, batch size is equal to 4

```
model.fit(x,y,epochs=2000,batch_size=4)
```

Let's check values of weights after training

```
print(model.get_weights());
```

Now, let's see our model's predictions

```
model.predict(x,batch_size=4)
```
